1. Write a decorator `n_times` which given a number `n` makes a function to execute `n` times.


```python
@n_times(3)
def print_my_name():
    print('Due')

print_my_name()

terminal
Due
Due
Due ```

````python
import functools

def n_times(number: int):
    def decorator(func):
        @functools.wraps(func)
        def inner(*args, **kwargs):
            for _ in range(number):
                func(*args, **kwargs)
            
        return inner
    return decorator
````

````python
@n_times(3)
def print_my_name():
    print('Due')

print_my_name()
````

2. Write a decorator `once` which makes a function to execute only once and other times it just returns the result calculated on the first call.

```python
@once
def random_number():
    return random.randint(1, 10)

print(random_number())
print(random_number())
print(random_number())

terminal
4
4
4

```

````python
def once(func):

    has_executed = False
    result = 0
    @functools.wraps(func)
    def inner(*args, **kwargs):
        nonlocal result, has_executed
        if not has_executed:
            result = func(*args, **kwargs)
            has_executed = True
        return result
        
    return inner
````

````python
import random
@once
def random_number():
    return random.randint(1, 10)

print(random_number())
print(random_number())
print(random_number())
````
